**Были проверены статистически гипотезы:**

Вывод: При значении p ниже 0,05 мы можем сказать, что разница в пропорциях двух выборок статистически значима. Следовательно, количество отточных клиентов с кредитными картами больше.

## Проверка статистических гипотез:
<a id="hepotesis"></a>

### Проверить равнозначность среднего заработка уходящих не уходящих клиентов:
<a id="hep2"></a>

Для проверки t-теста, важно учесть основные правила:

1) выборки достаточно велики

2) выборки независимы друг от друга

3) выборки не имеют больших различий по размеру

```python
churn_0 = data[data['churn']==0]['est_salary']
churn_1 =  data[data['churn']==1]['est_salary']
print('Количество остающихся клиентов:',len(churn_0))
print('Количество уходящих клеинтов:',len(churn_1))
```

```python
churn_0.hist(bins = 30, alpha=0.5)
churn_1.hist(bins = 30, alpha=0.5);
```

По графику можно увидеть нормальное распределение каждой выборки, которая не зависит друг от друга, однако выборки по размеру имеют большую разницу, поэтому для t-теста параметра equal_var укажем False 

Нулевая гипотеза: Средняя ЗП уходящих клиентов, равна средней ЗП остающихся клиентов.

Альтернативная гипотеза: Средняя ЗП уходящих клиентов, не равна средней ЗП остающихся клиентов.

Если p-value меньше порогового значения alpha, то гипотезу отвергаем. Иначе - не отвергаем.

```python
# results = вызов метода для проверки гипотезы

results = st.ttest_ind (
    data[data['churn']==0]['est_salary'],
    data[data['churn']==1]['est_salary'], equal_var = False)

alpha = 0.05

# вывод значения p-value на экран 
print (results.pvalue) 
# условный оператор с выводом строки с ответом

if results.pvalue < alpha:
    print ('Отвергаем нулевую гипотезу')
else:
    print ('Не получилось отвергнуть нулевую гипотезу')
```

```python
 print(data[data['churn']==0]['est_salary'])
```

t-тест вывел, что нельзя отвергнуть нулевую гипотезу, следовательно нет весомой разницы у доходов отточных клиентов и отающихся.

Проверим те же гипотезы, но u-тестом, где главная разница с t-тестом, выборки могут иметь выбросы, благодаря данному тесту, можно лучше определить статистическую значимость выборок и более точно ответить на наши гипотезы.

```python
alpha = 0.05  # критический уровень статистической значимости

results = st.mannwhitneyu(data[data['churn']==0]['est_salary'],\
                          data[data['churn']==1]['est_salary'])

print('p-значение: ', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу: разница статистически значима')
else:
    print(
        'Не получилось отвергнуть нулевую гипотезу, вывод о различии сделать нельзя'
    ) 
```

u-тест вывел, что отвергаем нулевую гипотезу, то есть между отточными клиентыми и остающимся разница в доходе есть.

Так как мы видим не нормальное распределение, лучше положиться на u-тест, так t-тест не имеет релевантность при не нормальной распределении.

Поэтому я выбираю итог u-теста.

Так как t-критерий является параметрическим, в его условия входит, что выборка должна быть распределена нормально. Для случаев, когда это не происходит и сущесвует u-критерий, который является непараметрическим.

Нулевая гипотеза: выборка распределена нормально

Альтернативная гипотеза: выборка распределена не нормально.

Если p-значение меньше 0.05 мы отвергаем нулевую гипотезу, иначе оставляем.


```python
#Проверим две выборки на распределение:
shapiro(data[data['churn']==0]['est_salary'])
```

### Проверить равнозначность наличие кредитной карты уходящих и не уходящих клиентов:
<a id="hep3"></a>

```python
#Выявим два датафрейма, где будут клиенты оттока 
credit_card_0 = data.query('churn == 1 and credit_card==0')

credit_card_1 = data.query('churn == 1 and credit_card==1')

print('Количество отточных клиентов без кредитной карты:', credit_card_0['userid'].count())
print('Количество отточных клиентов с кредитной картой:', credit_card_1['userid'].count())
```

Нулевая гипотеза: Средние количество кредитных карт  клиентов, равен среднему не наличию количеству кредитных карт клиентов.

Альтернативная гипотеза: Средние количество кредитных карт клиентов, не равен среднему не наличию количеству кредитных карт клиентов.

Если p-value меньше порогового значения alpha, то гипотезу отвергаем. Иначе - не отвергаем.

Применим z-тест

```python
from statsmodels.stats.proportion import proportions_ztest
successes = np.array([817, 1003]) # успехи
nobs = np.array([10000, 10000]) # попытки

#Apply z-test:
stat, pval = proportions_ztest(successes, nobs)

#Print results:
print('The z-value is: ', stat)
print('The p-value is: ', pval)
```

При значении p ниже 0,05 мы можем сказать, что разница в пропорциях двух выборок статистически значима.
Следовательно, количество отточных клиентов с кредитными картами больше. 

```python
def do_proportions_ztest(data, context:str, success:str, alpha:float = .05) -> None:
    '''Функция делает проверку на доли на заданном датасете'''
    USERID = 'userid'
    print(f'Нулевая гипотеза: среди клиентов в разрезе признака {context} доли имеющих признак {success} равны')
    print(f'Альтернативная гипотеза: среди клиентов в разрезе признака {context} доли имеющих признак {success} отличаются')

    nobs = data.groupby(context).agg({USERID:'count'})[USERID] # попытки
    successes = data.query(f'{success} == 1').groupby(context).agg({USERID:'count'})[USERID] # успехи

    _, pvalue = proportions_ztest(successes, nobs)
    
    print(f'p-value: {pvalue:.2e}' if pvalue < 1e-3 else f'p-value: {pvalue:.2%}')
    print('Отвергаем нулевую гипотезу' if pvalue < alpha else 'Принимаем нулевую гипотезу')
    print('-'*30)
    
do_proportions_ztest(data,'churn','credit_card')
do_proportions_ztest(data,'credit_card','churn')
```
